## Apache Access Log Analysis:

```python
# Parsed the log data into pandas dataframe
```

```python
import pandas as pd

df = pd.read_csv(
    'access.log',
    sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
    engine='python',
    na_values='-',
    header=None,
    usecols=[0, 3, 4, 5, 6, 7, 8],
    names=['ip', 'time', 'request', 'status', 'size', 'referer', 'user_agent'])
```

```python
pd.set_option('display.max_columns',156)
display(df)
```

## Data Cleaning

```python
df['referer'] = df['referer'].replace('"-"','Null')
df.head()
```

```python
df.info()
```

```python
df['time'] = df['time'].str.lstrip('[')
df['time'] = df['time'].str.rstrip(']')
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'],format="%d/%b/%Y:%H:%M:%S %z",errors='coerce')
df.info()
```

```python
df['time'] = pd.to_datetime(df['time'],format="%d/%m/%Y:%H:%M:%S.%f %z",errors='coerce')
df.info()
```

## Detecting for bots in the access log

```python
# Checking for some common found bots
bot_user_agents = ['Googebot','Bingbot','Applebot']
df['is_bot'] = df['user_agent'].str.contains('|'.join(bot_user_agents),case=False)
```

## Proportion of normal status and Error status code

```python
df['status'] =df['status'].astype(str)
normal_status_proportion = (df['status'].str.startswith('2').sum()) / len(df)
error_status_proportion = (df['status'].str.startswith(('4', '5')).sum()) / len(df)
print("Proportion of Normal status code: ",normal_status_proportion)
print("Proportion of Error status code: ",error_status_proportion)
```
