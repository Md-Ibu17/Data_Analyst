```python
import pandas as pd

df = pd.read_csv(
    'access.log',
    sep=r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
    engine='python',
    na_values='-',
    header=None,
    usecols=[0, 3, 4, 5, 6, 7, 8],
    names=['ip', 'time', 'request', 'status', 'size', 'referer', 'user_agent'])
```

```python
pd.set_option('display.max_columns',156)
display(df)
```

```python
df['referer'] = df['referer'].replace('"-"','Null')
df.head()
```

```python
df.info()
```

```python
df['time'] = df['time'].str.lstrip('[')
df['time'] = df['time'].str.rstrip(']')
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'],format="%d/%b/%Y:%H:%M:%S %z",errors='coerce')
df.info()
```

```python
df['time'] = pd.to_datetime(df['time'],format="%d/%m/%Y:%H:%M:%S.%f %z",errors='coerce')
df.info()
```

```python
df['request'] = df['request'].str.strip('"')
df[['method','url','Http']] = df['request'].str.split(" ",expand=True)
df.head()
```

```python
df['user_agent'] = df['user_agent'].str.strip('"')
df.head()
```
