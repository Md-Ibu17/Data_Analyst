```python
# Data Visualization
# Restaurants delivering Online or not
a=df['online_order'].value_counts()
a
```

```python
a.plot(kind='bar')
plt.xlabel('availability')
plt.ylabel('count')
plt.title('Online order availability')
```

```python
# Restaurants book table 
b=df['book_table'].value_counts()
b.plot(kind='bar')
plt.xlabel('availability')
plt.ylabel('count')
plt.title('Table booking availability')
```

```python
# Table booking Rate vs Rate
b=df['rating'].value_counts()
b.plot(kind='line')
plt.xlabel('Rating')
plt.ylabel('count')
plt.title('Table booking rating')
```
