## Data Cleaning

Definition: 
    Data cleaning is the process of fixing or removing incorrect, corrupted, incorrectly formatted, duplicate, or incomplete data within a dataset.
    
Removing
1) Duplicates
2) incorrect data
3) incorrect format
4) corrupted data 
5) incomplete data 
in the dataset

```python
import pandas as pd
```

```python
# customer call list
df = pd.read_excel(r"E:\Data Analyst\pandas learning\Customer Call List.xlsx")
df
```

We are gonna clean the data and change it into a usefull info.

```python
# Removing duplicates from the dataset
df = df.drop_duplicates()
df
```

This removes the duplicate ie) the last 2 rows were the same data

```python
# Removing the columns that we don't need (ie) The data's we are not gonna use
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# The strip() removes the whitespaces 
# note: We can also name lstrip for left and rstrip for right side strip 
df['Last_Name'].str.strip()
```

```python
# We need to remove the uncessary things in the lastname other than the name(str)
df['Last_Name'].str.lstrip("...")
df['Last_Name'].str.lstrip("/")
df['Last_Name'].str.rstrip("_")
```

```python
# Or You can do the above in a single line 
df['Last_Name'].str.strip("./_")
```

```python
#In order to save the df u need to do as follows
df['Last_Name']=df['Last_Name'].str.lstrip("...")
df
```

```python
df
```

Now cleaning the phone number column.  
We can't able to manually clean the whole column invdividual so we to automate this (what if there are around 20000 data)

```python
# First we need to remove all the symbols,letters from the phone number column by using replace()
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df
```

##Funtion replace()    
replace(a,b)    
where a is the thing u want to replace     
where b is the value replaced with 

```python
# The format of the phone number is 897-764-3426 this
# So now we need to add - after every 3 digits 
```

```python
# This can be down by index slicing but for that it need to a str 
# So first we are gonna convert the column individual to str (as followes)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
df['Phone_Number']
```

```python
# We can use lamba or for loop 
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number']
```

```python
# finally we removing the nan-- to a empty str 
df['Phone_Number'] = df['Phone_Number'].str.replace("nan--","")
df['Phone_Number'] = df['Phone_Number'].str.replace("Na--","")
df['Phone_Number']
```

```python
df
```

```python
# cleaning address column 
df["Address"].str.split(',',expand=True)
```

```python
# We are gonna give name for the seperated column 
df[["Street Adress","State","zip code"]] = df["Address"].str.split(',',expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
# Need to removes those "N/a" and Nan 
df = df.replace('N/a','')
df
```

```python
# The "NaN" can't be removed using relace() because its a value which means empty 
# So what we gonna do is use fillna() - fillname to change it
df = df.fillna('')
df
```

## Now we have cleaned the data but our objective is to give a customer list whome they can call 

```python
# Removing all the data which has no number as well as not contactable as given in the column "Do_Not_Contact"
```

```python
# Removing the rows whose Do_Not_Contact column says Y (filtering the customer list)
for i in df.index:
    if df.loc[i,'Do_Not_Contact'] == 'Y':
        df.drop(i,inplace=True)
df
```

```python
# Removing the rows whose phonenumber column is empty
for i in df.index:
    if df.loc[i,'Phone_Number'] == '':
        df.drop(i,inplace=True)
df
#Another way to drop null values
# df = df.dropna(subset = 'Phone_Number',inplace =True)
```

```python
# Finally we need to reset the index
df.reset_index()
```
